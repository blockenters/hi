```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('ecommerce.csv', index_col=0)
```

```python
df
```

```python
df.isna().sum()
```

```python
y = df['Sales']
```

```python
X = df.iloc[ : ,  :  -2+1  ]
```

```python
y
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=27)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor2 = RandomForestRegressor( random_state=27, n_estimators=500 )
```

```python
regressor2.fit(X_train, y_train)
```

```python
y_pred2 = regressor2.predict(X_test)
```

```python
mean_squared_error(y_test, y_pred2)
```

```python
r2_score(y_test, y_pred2)
```

```python
df = pd.read_csv('healthcare.csv', index_col=0)
```

```python
df
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df
```

```python
y = df['InsuranceClaim']
```

```python
X = df.iloc[ : , : -2+1 ]
```

```python
X['Gender'].unique()
```

```python
# Gender 레이블인코딩. 
```

```python
# 과일
# 사과, 파인애플, 오렌지, 자몽 
#   0,  1, 2, 3
```

```python
# Smkoker 는 레이블인코딩
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
encoder_gender = LabelEncoder()
```

```python
X['Gender'] = encoder_gender.fit_transform(X['Gender'])
```

```python
X
```

```python
encoder_smoker = LabelEncoder()
```

```python
X['Smoker'] = encoder_smoker.fit_transform( X['Smoker']  )
```

```python
X
```

```python
ct = ColumnTransformer( [('encoder', OneHotEncoder(), [3])] , remainder='passthrough' )
```

```python
# ['East', 'North', 'South', 'West']
```

```python
X = ct.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=27)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
regressor2 = RandomForestRegressor(random_state=27, n_estimators=500)
```

```python
regressor2.fit(X_train, y_train)
```

```python
y_pred2 = regressor2.predict(X_test)
```

```python
mean_squared_error(y_test, y_pred2)
```

```python
r2_score(y_test, y_pred2)
```
